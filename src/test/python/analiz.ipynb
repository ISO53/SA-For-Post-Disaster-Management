```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/termi/Documents/Projects/SchoolProjects/SA-For-Post-Disaster-Management/src/test/python/mutation_data.csv", delimiter=',')
data_2 = pd.read_csv("C:/Users/termi/Documents/Projects/SchoolProjects/SA-For-Post-Disaster-Management/src/test/python/mutation_data_2.csv", delimiter=',')
```

```python
# Assuming 'data' and 'data_2' are your datasets
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot boxplots for data
data.boxplot(grid=False, ax=axes[0])
axes[0].set_title('Data')

# Plot boxplots for data_2
data_2.boxplot(grid=False, ax=axes[1])
axes[1].set_title('Data_2')

plt.tight_layout()
plt.show()
```

```python
Results = pd.melt(data,var_name='Type', value_name='Score')
Results = Results.reset_index(drop=True)
Results['Type']= Results['Type'].astype('category')
```

```python
# Extract algorithm names
algorithm_names = list(data.columns)

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot heatmap for data
heatmap = axes[0].imshow(data.values, cmap='coolwarm', aspect='auto')
axes[0].set_title('Data')
axes[0].set_xticks(np.arange(len(algorithm_names)))
axes[0].set_xticklabels(algorithm_names)
fig.colorbar(heatmap, ax=axes[0], label='Fitness')

# Plot heatmap for data_2
heatmap = axes[1].imshow(data_2.values, cmap='coolwarm', aspect='auto')
axes[1].set_title('Data_2')
axes[1].set_xticks(np.arange(len(algorithm_names)))
axes[1].set_xticklabels(algorithm_names)
fig.colorbar(heatmap, ax=axes[1], label='Fitness')

plt.tight_layout()
plt.show()
```
